```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
City_A = [500000, 550000, 600000, 650000, 700000, 750000, 800000]
City_B = [800000, 850000, 900000, 950000, 1000000, 1050000, 1100000]
City_C= [1000000, 1050000, 1100000, 1150000, 1200000, 1250000, 1300000]
City_D=[1200000, 1250000, 1300000, 1350000, 1400000, 1450000, 1500000]
```

```python
plt.plot(years,City_A,marker='h',label="City A",linestyle=':')
plt.plot(years,City_B,marker='p',label="City B",linestyle=':')
plt.plot(years,City_C,marker='d',label="City C",linestyle=':')
plt.plot(years,City_D,marker='^',label="City D",linestyle=':')
plt.xlabel("years")
plt.ylabel("Population")
plt.title("Population Growth of Cities Over Time")
plt.legend()
plt.grid()
plt.show()
```
